```python
# parkir_tools.py

class TiketParkir:
    def __init__(self, jam_masuk):
        self.jam_masuk = jam_masuk
        self.tarif_per_jam = 2000

    def hitung_total(self, jam_keluar):
        # Menghitung selisih jam
        durasi = jam_keluar - self.jam_masuk

        # Antisipasi jika input jam keluar lebih kecil atau sama dengan masuk
        if durasi <= 0:
            durasi = 1

        total_bayar = durasi * self.tarif_per_jam
        return durasi, total_bayar
```
